```python
# This program takes a csv table of objects with the attributes: timeframe, center of object, bounding box;
# and a binary mask of a cell spreading to calculate how long the objects have stayed under the cell
# until they were ripped off/vanished.
# The output is a new csv table which shows the (normalized) amount of particles which were ripped off in each
# timestep (this differs from the timeframe in that it shows how long the particle was under, not the actual time)
# It does this for any csv table (with the right name) in the given folder.
# Additional output: Timeseries with leaving particles marked, Comparison to Analysis if just one mask is considered.
# Written by Eva Weber 
# Used for publication: "Forces during cellular uptake of viruses and nanoparticles at the ventral side" Wiegand et al. 2019

import csv
import numpy as np
from numpy import genfromtxt
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import imageio
import os
from glob import iglob
from skimage import io
import sys
from numpy.random import normal

sys._enablelegacywindowsfsencoding()

#NO BIG DIFFERENCE IN AMOUNT OF PARTICLES WHEN OUT OF FOCUS

# PUT YOUR FILENAMES/PATHS AND PARAMETERS HERE

slicedataset = 0       # only unequal zero, if only part of longer measurement is used
pathend = 'crgdAuNPs100persurfovernight'         # seperate from rest; names output
pathway = 'C:\\Users\\' + '\\' + pathend
make_array = 0               #for differentiation: turns to 1 after first initiation of array


for pathname in iglob(pathway + '\cell*_f_crop_CSV-Table.csv'): # loops over all csv tables, rename if necessary
    indexpathway = pathname.rfind('cell')             # finds last 'cell', gets number, finds other files
    numb_cell = pathname[indexpathway+4] + pathname[indexpathway+5]
    
    file1 = pathway + '\cell' + numb_cell + '_f_crop_CSV-Table.csv'         # FILE OF TABLE AFTER ILASTIK WITH PATH
    file2 = pathway + '\cell' + numb_cell + 'MaskStack.tif'                 # FILE OF MASKS (PNG)
    file2a = pathway + '\cell' + numb_cell + 'maskall.png'                  # all masks combined for first iteration
    file3 = pathway + '\cell' + numb_cell + '_f_crop.tif'                   # original data (registered and cropped)
    
    time_delay = 30                     # PUT TIME FROM SEEDING OF CELLS TO MEASURING
    time_step = 5 
    first_delay = 1              # generalization how long the particles are under cell in beginning in timesteps
    
    path_results = pathway + '\\mask_results_rel'

    if os.path.exists(path_results + numb_cell) != True:
        os.mkdir(path_results + numb_cell)

        
        
    mask_time = io.imread(file2)                    # read masks to array
    immask = imageio.imread(file2a)                 # combined mask for counting
    print(mask_time[0])
    in_cell = 255                                   # FOR IMMASK OUTSIDE; FOR MASK_TIME INSIDE




    my_data = genfromtxt(file1, delimiter=',')                        # read table to array

    line1_list = list(pd.read_csv(file1, nrows=0))                    # get first row, make list out of it

    # The table from Ilastik has more than the necessary information. 
    # The following parts are used in the program and the info is concentrated into a smaller array
    
    index_list = ['frame', 'Center_of_the_object_0', 'Center_of_the_object_1', 'Bounding_Box_Minimum_0', 
                  'Bounding_Box_Minimum_1', 'Bounding_Box_Maximum_0', 'Bounding_Box_Maximum_1'] 

    i = 0
    for name in index_list:                     # use names to get needed array slices in new array
        temphold = line1_list.index(name)
        tempdata = my_data[:,temphold]
        tempdata1 = np.delete(tempdata, 0)
        if i == 0:
            dataset = tempdata1
        else:
            dataset = np.vstack([dataset, tempdata1])
        i = i+1
    i = 0
    
    del my_data
    
    if slicedataset != 0:
        dataset = dataset[:, 0:np.argwhere(dataset[0] == slicedataset).max()]
    
    
    lasttrack = int(np.amax(dataset[0]))            # get last track (for exceptions and stopping points)
    total_part = 0                                  # total amount for total mask

    
    # check that mask is same size as pic
    if np.amax(dataset[4])>immask.shape[0] or np.amax(dataset[3])>immask.shape[1]: 
        print('MASK SMALLER THAN IMAGE')
        exit()


    'NEXT PART: IF NUMBER OF OBJECTS DRAMATICALLY DIFFERENT FROM ONE FRAME TO NEXT, IT IS TO BE IGNORED'
    'IF NUMBER OF PARTICLES RIPPED OFF IN THE BACKGROUND DRAMATICALLY DIFFERENT, FRAME IS TO BE IGNORED'
    
    # counts total amount of particles under cell 
    while dataset[0][i]<1:                          # make list from first frame trackIds
        if immask[int(dataset[2][i])][int(dataset[1][i])] != in_cell:
            total_part += 1
        i = i+1
    i = 0


    # Compares bounding boxes and distance of centers
    # If no box (or close center) is found in one loop, count as ripped off
    # COMPARES TO FIRST BOUNDING BOX ALWAYS!!!
    
    dist_cent = 3

    # Initiation of array for different calculations on torn off particles
    
    ripoffs = np.zeros((lasttrack+first_delay+1, 2))             # how many per timestep
    ripoffs_left = np.zeros((lasttrack+first_delay+1, 2))        # how many left per timestep
    ripoffs_xy = np.zeros((1, 3))              # get position of ripped off particle + timestep
    rip_true = np.zeros((lasttrack+first_delay+1, 2))      # values later with masks
    left_rel = np.zeros((lasttrack+first_delay+1, 2))     # left when relatively calculated
    calc_total = np.zeros((lasttrack+first_delay+1, 2))   # array of totals (multiple mask calculation)
    avail_total = np.zeros((lasttrack+first_delay+1, 2))  # amount of particles which could have been ripped off
    rip_min_beg = np.zeros((lasttrack+first_delay+1, 2)) # doesn't count those from beginning (under first mask), 
                                                         # counts everything for length of force impact
    left_abs = np.zeros((lasttrack+first_delay+1, 2))      # +1 to add 5 min in beginning
    stay_part = np.zeros((1, 2))               # get positions of permanent particles
    # amount of particles which stayed a certain amount for a certain time
    time_stay = np.zeros((lasttrack+first_delay+1, 2))     
    # border for values of different cells, one index longer because of titles
    commas_fl = np.zeros((lasttrack+first_delay+2, 1))     
    commas = commas_fl.astype(str)
    
    for i in range(0, lasttrack+first_delay+1):
        ripoffs[i][0] = i
        ripoffs_left[i][0] = i
        time_stay[i][0] = i
        commas[i][0] = ','
        rip_true[i][0] = i
        left_abs[i][0] = i
        rip_min_beg[i][0] = i
        left_rel[i][0] = i
        calc_total[i][0] = i
        avail_total[i][0] = i
    commas[lasttrack+first_delay+1][0] = ','
    
    plateau = 0    
    
    # Goes to every particle visible in the first frame and checks if it is ever under any mask
    # if yes, saves the timepoint when it is no longer found
    # (or the timeseries ends) and goes on to the next particle
    
    print('Particle analysis starting')
    i = 0
    while i <= np.argwhere(dataset[0] == 0).max():
        loop_j = int(np.argwhere(dataset[0] == 0).max())+1 # goes straight to frame one
        
        if immask[int(dataset[2][i])][int(dataset[1][i])] != in_cell: # remember: here outside of cell
            while loop_j < dataset[0].size:

                box_j_0 = np.arange(int(dataset[3][loop_j]), int(dataset[5][loop_j])+1) # make array with values 0
                box_j_1 = np.arange(int(dataset[4][loop_j]), int(dataset[6][loop_j])+1) # make array with values 1

                # see if any element is in box (as they are two boxes checking like this equals checking each)
                # see if any element is close to center of object (important for small boxes/weak particles)
                if ((np.any(box_j_0 > dataset[3][i]) and np.any(box_j_0 < dataset[5][i]) and 
                    np.any(box_j_1 > dataset[4][i]) and np.any(box_j_1 < dataset[6][i])) or 
                    (abs(dataset[1][loop_j]-dataset[1][i]) < dist_cent and 
                     abs(dataset[2][loop_j]-dataset[2][i]) < dist_cent)):
                    # if one there go to next frame
                    if dataset[0][loop_j] < lasttrack:
                        loop_j = int(np.argwhere(dataset[0] == (dataset[0][loop_j] + 1)).min())
                    # count staying particles for relative comparison
                    else:
                        stay_part = np.vstack([stay_part, [dataset[1][i], dataset[2][i]]])
                        break
                    continue
                if loop_j+1 >= dataset[0].size:
                    frame_rip = int(dataset[0][loop_j])
                    ripoffs[frame_rip][1] += 1
                    ripoffs_xy = np.vstack([ripoffs_xy, [dataset[0][loop_j]-1, dataset[1][i], dataset[2][i]]])

                    break
                if dataset[0][loop_j+1] != dataset[0][loop_j]: # if not found new box until next frame: count
                    if dataset[0][loop_j] == 1:   # if gone in first frame: count to plateau
                        plateau += 1
                    elif dataset[0][loop_j] != 1:    # other than first frame: count as torn off
                        frame_rip = int(dataset[0][loop_j])
                        ripoffs[frame_rip][1] += 1
                        ripoffs_xy = np.vstack([ripoffs_xy, [dataset[0][loop_j]-1, dataset[1][i], dataset[2][i]]])
                    break

                loop_j += 1
        i += 1
        
    ripoffs_xy = np.delete(ripoffs_xy, (0), axis=0) #delete initialization row
    stay_part = np.delete(stay_part, (0), axis=0)
    ripped_all = np.sum(ripoffs[:,1])
    ripoffs_left_pro = np.copy(ripoffs_left)
    total_part = total_part - plateau   # takes of particles which were ripped off/moved in first frame
    decr_part = total_part
    
    # Puts leftover particles in array (one whole mask), absolute and relative numbers
    for i in range(0, lasttrack+1):
        decr_part -= ripoffs[i][1]
        ripoffs_left[i][1] = decr_part
        ripoffs_left_pro[i][1] = decr_part/total_part

    export_csv = np.copy(ripoffs)
    export_csv = np.column_stack((export_csv, ripoffs_left[:,1]))
    export_csv = np.column_stack((export_csv, ripoffs_left_pro[:,1]))

    # exports array
    np.savetxt(path_results + numb_cell + '\\csv_without_masks' + '.csv', export_csv, delimiter=',')
    
    with open(path_results + numb_cell + '\\numbresults' + numb_cell + '.txt', 'w') as f:
            f.write('%s' % 'ripped_all')
            f.write('%d' % ripped_all)
            f.write('%s' % 'Plateau(ripped off in first step)')
            f.write('%d' % plateau)

    
    # Marks particles that vanish in the next frame with a cross (for visual control)
    images = io.imread(file3)
    
    brighter = 100
    w, h = images.shape[0], images.shape[1]
    for tracker in range(0, lasttrack):
        data = images[tracker]
        crosses = np.where(ripoffs_xy[:,0] == tracker)     # gets positions where crosses are made
        crosses = crosses[0]
        
        for i in range(0, len(crosses)):
            data[int(ripoffs_xy[crosses[i]][2])][int(ripoffs_xy[crosses[i]][1])] = brighter
            for k in range(0, 5):                     # extras to not move over image bounds
                if int(ripoffs_xy[crosses[i]][2])+k < data.shape[0]:
                    data[int(ripoffs_xy[crosses[i]][2])+k][int(ripoffs_xy[crosses[i]][1])] = brighter
                if int(ripoffs_xy[crosses[i]][2])-k > 0:
                    data[int(ripoffs_xy[crosses[i]][2])-k][int(ripoffs_xy[crosses[i]][1])] = brighter
                if int(ripoffs_xy[crosses[i]][1])+k < data.shape[1]:
                    data[int(ripoffs_xy[crosses[i]][2])][int(ripoffs_xy[crosses[i]][1])+k] = brighter
                if int(ripoffs_xy[crosses[i]][1])-k > 0:
                    data[int(ripoffs_xy[crosses[i]][2])][int(ripoffs_xy[crosses[i]][1])-k] = brighter
        img = Image.fromarray(data)
        img.save(path_results + numb_cell + '\\tryouts' + str(tracker) + '.png')
    io.imsave(path_results + numb_cell + '\\tryastack' + '.png', images)
    plt.close('all')



    # Here we look at the particles and figure out 
    # how long they have stayed under the cell (to calculate interaction time)
    for i in range(0, ripoffs_xy.shape[0]): # LOOK AT TIME+-1???? differentiate between two different possibilities
                                            # either mask moved away or stays on top
        if (mask_time[int(ripoffs_xy[i][0])+1][int(ripoffs_xy[i][2])][int(ripoffs_xy[i][1])] != in_cell and 
            mask_time[int(ripoffs_xy[i][0])][int(ripoffs_xy[i][2])][int(ripoffs_xy[i][1])] == in_cell):
            j = int(ripoffs_xy[i][0])
            
            force_time = 0                  # counts how long particle has been under mask
            
            while(mask_time[j][int(ripoffs_xy[i][2])][int(ripoffs_xy[i][1])] == in_cell):
                j -=1
                force_time += 1 
                if j == 0:
                    force_time += first_delay
                    break
            if j != 0:   # only counts if not the whole time under mask
                rip_min_beg[force_time][1] += 1
            rip_true[force_time][1] += 1
        elif mask_time[int(ripoffs_xy[i][0])+1][int(ripoffs_xy[i][2])][int(ripoffs_xy[i][1])] == in_cell: 
            j = int(ripoffs_xy[i][0])
            force_time = 0
            while(mask_time[j+1][int(ripoffs_xy[i][2])][int(ripoffs_xy[i][1])] == in_cell):
                j -=1
                force_time += 1
                if j == 0:
                    force_time += first_delay
                    break
            if j != 0:      # only counts if not the whole time under mask, as here no definite time frame
                rip_min_beg[force_time][1] += 1
            rip_true[force_time][1] += 1
        else:
            continue

            
            
    # Here we look at how long those particles that did not get torn off
    # have interacted with the cell
    for i in range(0, stay_part.shape[0]):
        track_time = 0
        j=0
        for j in range(0, lasttrack+1): # puts length of stay under cell in bins
            if mask_time[j][int(stay_part[i][1])][int(stay_part[i][0])] == in_cell and j == 0:
                track_time += first_delay
                continue
            elif mask_time[j][int(stay_part[i][1])][int(stay_part[i][0])] == in_cell and j != 0:
                track_time += 1
                if j == lasttrack:
                    time_stay[track_time][1] += 1
                    track_time = 0
                continue
            elif mask_time[j][int(stay_part[i][1])][int(stay_part[i][0])] != in_cell and track_time != 0:
                time_stay[track_time][1] += 1
                track_time = 0
                continue
                
    print(rip_true)
    left_over = np.copy(rip_true)
    left_min_beg = np.copy(rip_min_beg)
    sum_left = 0
    
    # In the next part, the program calculates the percentage of particles ripped off
    # Since the amount of particles that could be ripped off changes with each timestep,
    # we can't use one value. Instead, the amount it gets normalized to changes for each timestep
    # and is made up from the amount of particles that have interacted with the cell for AT LEAST
    # the same time as the current step, plus the amount that has been ripped off in previous
    # timesteps (in percent) as would be the case with the normalization with just the 
    # whole amount in a "normal" calculation
    
    for i in range(0, lasttrack+first_delay+1):
        sum_stay = time_stay[i:lasttrack+first_delay+2, 1] # how many particles which didn't get ripped off
        sum_go = rip_true[i:lasttrack+first_delay+2, 1]    # how many particles got ripped off (including this round)
        sum_all = np.sum(sum_stay) + np.sum(sum_go)
        avail_total[i][1] = sum_all
        if sum_left != 1:
            wprev_sum = sum_all/(1-sum_left)
            calc_total[i][1] = wprev_sum
            left_over[i][1] = rip_true[i][1]/wprev_sum
        # next: wprev_sum is technically wrong but can't divide by 0 and no particles should be left
        else:     
            wprev_sum = 0
            calc_total[i][1] = wprev_sum
            left_over[i][1] = 0
        sum_left = sum_left + left_over[i][1]           # percent of all gone particles
        left_rel[i][1] = 1-sum_left                     # percent of all particles left
    
        
    
    # make table of values for each cell and add to tables of previous cells
    # export all after for loop
    
    fhel = np.copy(rip_true[:,0])
    frame_times = fhel*time_step
    export_masks = np.copy(frame_times)
    export_masks = np.column_stack((export_masks, rip_true[:,1]))
    export_masks = np.column_stack((export_masks, avail_total[:,1]))
    export_masks = np.column_stack((export_masks, calc_total[:,1]))
    export_masks = np.column_stack((export_masks, left_over[:,1]))
    export_masks = np.column_stack((export_masks, left_rel[:,1]))
    export_masks = np.column_stack((export_masks, ripoffs_left_pro[:,1]))
    export_masks_str = export_masks.astype(str)
    titles_masks = np.array(['times', 'ripped off', 'available', 
                             'total', '% ripped off', '% stayed', 'left; one mask'])
    
    export_masks_str = np.insert(export_masks_str, 0, titles_masks, 0)
    
    if make_array == 0:
        export_array = export_masks_str
    else:
        export_array = np.column_stack((export_array, commas))
        export_array = np.concatenate((export_array, export_masks_str), axis=1)
    
    make_array = 1    
    
    # plot original analysis (one mask) and new analysis, plus the amount ripped off in each step
    plt.plot(left_over[:,0], left_over[:,1])
    plt.plot(left_rel[:,0], left_rel[:,1])
    plt.plot(ripoffs_left_pro[:,0], ripoffs_left_pro[:,1])
    plt.legend(['mult_masks', 'mult masks rel', 'one mask'], loc='upper right')
    plt.savefig(path_results + numb_cell + '\\plt_comp_diffandsamemask.png')
    plt.show()
    
# export csv table of values of all cells
with open(pathway + '\\csv_with_masks' + pathend + '.csv','wb') as f:
    np.savetxt(f,export_array,fmt='%s', delimiter=',')  
with open('C:\\Python\\results_from_multmasks_multtotals' 
          + '\\csv_with_masks' + pathend + '.csv','wb') as f:
    np.savetxt(f,export_array,fmt='%s', delimiter=',')  
```
